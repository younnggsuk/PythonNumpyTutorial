# Matploilib

Matplotlib는 plotting 라이브러리이다. 여기서는 `matplotlib.pyplot`모듈에 대해 간단히 소개한다.

## Plotting
`plot()`은 2D data로 plot해준다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

다음과 같이 한번에 여러개의 선을 plot할 수도 있으며, title, legend, axis label도 추가할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.title("Sine and Cosine")
plt.legend(['Sine', 'Cosine'])
plt.show()
```

`plot()`에 대한 자세한 내용은 [문서](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)를 참고하자.

## Subplots

`subplot()`을 사용하면, 같은 figure에 여러 그래프를 plot할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title("Sine")

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title("Cosine")

plt.show()
```

`subplot()`에 대한 자세한 내용은 [문서](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot)를 참고하자.

## Images

`imshow()`를 사용하면, 이미지를 보여줄 수 있다.

```python
import cv2
import numpy as np

img = cv2.imread("./cat.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_tinted = img * [1, 0.95, 0.9]
img_tinted = np.uint8(img_tinted)

plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(img_tinted)

plt.show()
```
